# 1 Monte-Carlo Simulation of firm's behavior in exportation

### 1. Setup a firm's problem.

In equilibrium we have:
production= demand
$$
\rightarrow Al^\alpha = C
$$

Domestic firm:
$$
C_0 = Y(\frac{p_0}{P})^{-\sigma} \Rightarrow l = (\frac{Y}{A})^{\frac{1}{\alpha}}(\frac{p_0}{P})^{\frac{-\sigma}{\alpha}}
$$
$$
\pi_0 = C_0p_0 - wl - f_0 = Y(\frac{p_0}{P})^{-\sigma}p_0 - w {(\frac{Y}{A})}^{\frac{1}{\alpha}}{(\frac{p_0}{P})}^{-\frac {\sigma}{\alpha}} - f_0
$$
Exporter firm:
$$
C_x = Y(\frac{p_x}{P})^{-\sigma}
$$
$$
Al^\alpha = C_d + C_x = Y P^\sigma (p_d^{-\sigma} + p_x^{-\sigma}) \Rightarrow l = (\frac{YP^\sigma}{A})^{\frac{1}{\alpha}} (p_d^{-\sigma} + p_x^{-\sigma})^{\frac{1}{\alpha}}
$$
$$
\pi_x = C_dp_d + C_x(p_x - \tau p_x) - wl -f_0 - f_x
$$
$$
 =  Y P^\sigma (p_d^{1-\sigma} + (1-\tau)p_x^{1-\sigma}) - w (\frac{YP^\sigma}{A})^{\frac{1}{\alpha}} (p_d^{-\sigma} + p_x^{-\sigma})^{\frac{1}{\alpha}} -f_0 - f_x
$$

The firm wants to maximize its profit, so the firm's problem is:
$$
Domestic: \max_{p_0}{(Y(\frac{p_0}{P})^{-\sigma}p_0 - w {(\frac{Y}{A})}^{\frac{1}{\alpha}}{(\frac{p_0}{P})}^{-\frac {\sigma}{\alpha}} - f_0)}
$$
$$
Exporter: \max_{p_d,\ p_x}{(Y P^\sigma (p_d^{1-\sigma} + (1-\tau)p_x^{1-\sigma}) - w (\frac{YP^\sigma}{A})^{\frac{1}{\alpha}} (p_d^{-\sigma} + p_x^{-\sigma})^{\frac{1}{\alpha}} -f_0 - f_x)}
$$

### 2. Solve for the the firm's price.

Domestic:
$$
\frac{\partial \pi_0}{\partial p_0} = 0 \Rightarrow (1-\sigma)YP^{\sigma} p_0^{- \sigma} + w(\frac{\sigma}{\alpha})(\frac{Y}{A})^{\frac {1}{\alpha}}P^{\frac{\sigma}{\alpha}}p_0^{(-\frac{\sigma}{\alpha} - 1)}=0
$$
$$
\Rightarrow p_0 = [\frac{w \frac{\sigma}{\alpha}(\frac{Y}{A})^{\frac{1}{\alpha}} P^{\frac{\sigma}{\alpha}}}{(\sigma -1)YP^{\sigma}}]^{\frac {1}{-\sigma + \frac {\sigma}{\alpha} +1} }
$$

Exporter:
$$
\frac{\partial \pi_x}{\partial p_d} = 0 \Rightarrow Y P^\sigma (1-\sigma) + \frac{w\sigma}{\alpha} (\frac{YP^\sigma}{A})^{\frac{1}{\alpha}} (p_d^{-\sigma} + p_x^{-\sigma})^{\frac{1-\alpha}{\alpha}} p_d^{-1}
$$
$$
\frac{\partial \pi_x}{\partial p_x} = 0 \Rightarrow Y P^\sigma (1-\sigma)(1-\tau) + \frac{w\sigma}{\alpha} (\frac{YP^\sigma}{A})^{\frac{1}{\alpha}} (p_d^{-\sigma} + p_x^{-\sigma})^{\frac{1-\alpha}{\alpha}} p_x^{-1}
$$

### 3. Determine the threshold ($A_0$) the a firm enters the domestic market.

For entering the domestic market, the $\pi_0$ should be positive:
$$
\pi_0 > 0 \Rightarrow Y(\frac{p_0}{P})^{-\sigma}p_0 - w {(\frac{Y}{A})}^{\frac{1}{\alpha}}{(\frac{p_0}{P})}^{-\frac {\sigma}{\alpha}} - f_0 > 0
$$
$$
\Rightarrow w (\frac{Y}{A})^{\frac{1}{\alpha}} (\frac{p_0}{P})^{-\frac {\sigma}{\alpha}} < Y(\frac{p_0}{P})^{-\sigma}p_0 - f_0
$$
$$
\Rightarrow (\frac{Y}{A})^{\frac{1}{\alpha}} < (\frac{p_0}{P})^{\frac {\sigma}{\alpha}}\frac{Y(\frac{p_0}{P})^{-\sigma}p_0 - f_0}{w}
$$
$$
\Rightarrow \frac{Y}{A} < [(\frac{p_0}{P})^{\frac {\sigma}{\alpha}}\frac{Y(\frac{p_0}{P})^{-\sigma}p_0 - f_0}{w}]^{\alpha}
$$
$$
\Rightarrow A_0 > Y(\frac{p_0}{P})^{-\sigma}[\frac{w}{Y(\frac{p_0}{P})^{-\sigma}p_0 - f_0}]^{\alpha}
$$

### 4. Determine the threshold ($A_x$) the a firm enters the export market.

For entering the export market, the $\pi_x$ should be greater than $\pi_0$:
$$
\pi_x > \pi_0 \Rightarrow Y P^\sigma (p_d^{1-\sigma} + (1-\tau)p_x^{1-\sigma}) - w (\frac{YP^\sigma}{A})^{\frac{1}{\alpha}} (p_d^{-\sigma} + p_x^{-\sigma})^{\frac{1}{\alpha}} -f_0 - f_x > Y(\frac{p_0}{P})^{-\sigma}p_0 - w (\frac{Y}{A})^{\frac{1}{\alpha}} (\frac{p_0}{P})^{-\frac {\sigma}{\alpha}} - f_0
$$
$$
\Rightarrow Y P^\sigma (p_d^{1-\sigma} + (1-\tau)p_x^{1-\sigma} - p_0^{1-\sigma}) - f_x > (\frac{1}{A})^{\frac{1}{\alpha}} w Y^{\frac{1}{\alpha}} P^{\frac{\sigma}{\alpha}}
\times [(p_d^{-\sigma} + p_x^{-\sigma})^{\frac{1}{\alpha}} - p_0^{\frac{-\sigma}{\alpha}}]
$$
$$
\Rightarrow (\frac{1}{A})^{\frac{1}{\alpha}} < \frac{Y P^\sigma (p_d^{1-\sigma} + (1-\tau)p_x^{1-\sigma} - p_0^{1-\sigma}) - f_x}{w Y^{\frac{1}{\alpha}} P^{\frac{\sigma}{\alpha}} \times [(p_d^{-\sigma} + p_x^{-\sigma})^{\frac{1}{\alpha}} - p_0^{\frac{-\sigma}{\alpha}}]}
$$
$$
\Rightarrow \frac{1}{A} < [\frac{Y P^\sigma (p_d^{1-\sigma} + (1-\tau)p_x^{1-\sigma} - p_0^{1-\sigma}) - f_x}{w Y^{\frac{1}{\alpha}} P^{\frac{\sigma}{\alpha}}
\times [(p_d^{-\sigma} + p_x^{-\sigma})^{\frac{1}{\alpha}} - p_0^{\frac{-\sigma}{\alpha}}]}]^\alpha
$$
$$
\Rightarrow A_x > [\frac{w Y^{\frac{1}{\alpha}} P^{\frac{\sigma}{\alpha}}
\times [(p_d^{-\sigma} + p_x^{-\sigma})^{\frac{1}{\alpha}} - p_0^{\frac{-\sigma}{\alpha}}]}{Y P^\sigma (p_d^{1-\sigma} + (1-\tau)p_x^{1-\sigma} - p_0^{1-\sigma}) - f_x}]^\alpha
$$

### 5. Now write a code in Python to simulate the firm's behavior using the Monte-Carlo Simulation.

```julia
using Distributions, Plots, Optim, Random, DataFrames, BlackBoxOptim
```

```julia
π₀(p₀, σ, P, Y, α, f₀, w, A) = Y * P^σ * p₀^(1-σ) - w * (Y * P^σ / A)^(1/α) * p₀^(-1 * σ / α) - f₀
πₓ(p, σ, P, Y, α, f₀, fₓ, τ, w, A) = Y * P^σ * (p[1]^(1-σ) + (1-τ) * p[2]^(1-σ)) - w * (Y * P^σ / A)^(1/α) * (sum(p.^(-σ)))^(1 / α) - f₀ - fₓ;
```

```julia
function firm_sim(θ, Ā, σ, P, Y, α, f₀, fₓ, w, τ, N, dist="Pareto")
    Random.seed!(1395)
    if dist == "Pareto"
        A = rand(Pareto(θ, Ā), N)
    elseif dist == "Lognormal"
        log_A = rand(Normal(Ā, θ), N)
        A = exp.(log_A)
    end
    p₀ = [] #Domestic price if the firm is only doemstic
    p₀ₓ = [] #Domestic price if the firm exports
    pₓ = [] #Export price
    for i ∈ A
        π₀_opt(p) = -π₀(p, σ, P, Y, α, f₀, w, i)
        p = optimize(π₀_opt, 0., 50.).minimizer
        append!(p₀, p)
        πₓ_opt(p) = -πₓ(p, σ, P, Y, α, f₀, fₓ, τ, w, i)
        p = optimize(πₓ_opt, [0.1, 0.1]).minimizer
        append!(p₀ₓ, [p[1]])
        append!(pₓ, [p[2]])
    end
    #Profits
    d_profit = π₀.(p₀, σ, P, Y, α, f₀, w, A)
    e_profit = [πₓ([p₀ₓ[i], pₓ[i]], σ, P, Y, α, f₀, fₓ, τ, w, j) for (i, j) ∈ enumerate(A)]
    firms = DataFrame()
    firms.domestic_profit = d_profit
    firms.export_profit = e_profit
    firms.operating = d_profit .> 0
    firms.Exporter = e_profit .> d_profit
    #Outputs
    d_output = Y .* (p₀./P).^(-σ)
    d_output[d_profit .< 0.] .= 0.
    d_output[e_profit .> d_profit] .= Y .* (p₀ₓ[e_profit .> d_profit]./P).^(-σ)
    firms.Domestic_Output = d_output
    e_output = Y .* (pₓ./P).^(-σ)
    e_output[e_profit .< d_profit] .= 0.
    firms.Export_Output = e_output
    firms.Total_Output = d_output .+ e_output
    #Labor
    labors = (firms.Total_Output ./ A).^(1/α)
    firms.Labors = labors
    #Sales
    domestic_sales = (1 .- firms.Exporter) .* p₀ .* d_output .+ firms.Exporter .* p₀ₓ .* d_output
    firms.domestic_sales = domestic_sales
    exports = pₓ .* e_output
    firms.exports = exports
    sales = domestic_sales .+ exports
    firms.sales = sales
    
    return firms
end
```

### 6. Suppose the number of firms is N =1000. Generate N samples of firms and save them in your memory.

```julia
N=1000
θ=3.5
Ā=1
σ=2.4
P=1
Y=100
α=.5
f₀=1.2
fₓ=4
w=1
τ = 0.1;
```

```julia
firms = firm_sim(θ, Ā, σ, P, Y, α, f₀, fₓ, w, τ, N)
```

### 8. SMM

Moments

```julia
m1 = mean(firms.sales[firms.operating .> 0])
m1 = isnan(m1) ? 0. : m1
m2 = mean(firms.exports[firms.exports .> 0.])
m2 = isnan(m2) ? 0. : m2
m3 = sum(firms.Exporter) / N
m3 = isnan(m3) ? 0. : m3
m4 = mean(firms.Labors[(firms.operating .+ firms.Exporter) .== 1])
m4 = isnan(m4) ? 0. : m4;
```

```julia
firm_sim(θ::Vector, Ā, P, Y, α, w, τ, N) = firm_sim(θ[1], Ā, θ[2], P, Y, α, θ[3], θ[4], w, τ, N)
function SMM(θ)
    firms = firm_sim(θ, Ā, P, Y, α, w, τ, N)
    average_sales = mean(firms.sales[firms.operating .> 0])
    average_sales = isnan(average_sales) ? 0. : average_sales
    average_export = mean(firms.exports[firms.exports .> 0.])
    average_export = isnan(average_export) ? 0. : average_export
    export_fraction = sum(firms.Exporter) / N
    export_fraction = isnan(export_fraction) ? 0. : export_fraction
    average_domestic_employment = mean(firms.Labors[(firms.operating .+ firms.Exporter) .== 1])
    average_domestic_employment = isnan(average_domestic_employment) ? 0. : average_domestic_employment
    smm = (m1 - average_sales)^2 + 
          (m2 - average_export)^2 + 
          (m3 - export_fraction)^2 + 
          (m4 - average_domestic_employment)^2
    return smm
end
```

```julia
θ_opt = optimize(SMM, [3, 2, 1, 4.]).minimizer
println("The estimation for θ is $(θ_opt[1])")
println("The estimation for σ is $(θ_opt[2])")
println("The estimation for f₀ is $(θ_opt[3])")
println("The estimation for fₓ is $(θ_opt[4])")
```

### 9. Now repeat the previous problem, but now you add two other moments and re-estimate the model.

```julia
function SMM2(θ)
    firms = firm_sim(θ, Ā, P, Y, α, w, τ, N)
    average_sales = mean(firms.sales[firms.operating .> 0])
    average_sales = isnan(average_sales) ? 0. : average_sales
    average_export = mean(firms.exports[firms.exports .> 0.])
    average_export = isnan(average_export) ? 0. : average_export
    export_fraction = sum(firms.Exporter) / N
    export_fraction = isnan(export_fraction) ? 0. : export_fraction
    average_domestic_employment = mean(firms.Labors[(firms.operating .+ firms.Exporter) .== 1])
    average_domestic_employment = isnan(average_domestic_employment) ? 0. : average_domestic_employment
    std_sales = std(firms.sales[firms.operating .> 0])
    std_sales = isnan(std_sales) ? 0. : std_sales
    std_employment = std(firms.Labors[(firms.operating .+ firms.Exporter) .== 1])
    std_employment = isnan(std_employment) ? 0. : std_employment
    smm = (m1 - average_sales)^2 + 
          (m2 - average_export)^2 + 
          (m3 - export_fraction)^2 + 
          (m4 - average_domestic_employment)^2 + 
          (m5 - std_sales)^2 + 
          (m6 - std_employment)^2
    return smm
end
```

```julia
θ_opt = optimize(SMM2, [3, 2, 1, 4.]).minimizer
println("The estimation for θ is $(θ_opt[1])")
println("The estimation for σ is $(θ_opt[2])")
println("The estimation for f₀ is $(θ_opt[3])")
println("The estimation for fₓ is $(θ_opt[4])")
```
